```python
def csqi_calc(signal,signal_samplingrate,curve,curvefreq):
    #function to calculate curve matching sqi
    #signal - single channel signal
    #signal_samplingrate - sampling frequency of the signal
    #  curve - curve against which correlation of each heartbeat section
    #is compared.
    # curvefreq - sampling rate of the curve


    if np.sum(signal)==0:
        sqi=np.empty((1, signal.shape[1]))
        c_sqi=np.empty((1, signal.shape[1]))
    else:
        Fs=signal_samplingrate;
        if signal.shape[0]!=1:
            signal=signal.transpose()
            signal=np.reshape(signal, (1, -1))
    
        sqi=np.zeros((1, signal.shape[1]))
        c_sqi=np.zeros((1, signal.shape[1]))
        if Fs!=curvefreq:
            curve = scipy.signal.resample_poly(curve,Fs,curvefreq);
        window_size=len(curve);
        signal2=np.hstack((np.fliplr(np.reshape(signal[0,0:round(window_size/2)],(1,-1))),signal,np.fliplr(np.reshape(signal[0,(len(signal)-round(window_size/2)):signal.shape[1]],(1,-1)))));#mirroring signal towards the beginning and end

        k=0;
        x=curve.transpose()
        y=np.fliplr(x)
        y=np.hstack((y,np.zeros((1,window_size-1))))
        P=circulant(y[0,:])
        P=P.transpose()
        P=P[0:window_size,:]
        lag=np.arange(-(window_size-1),window_size)
    
    
        #window=np.hstack((np.reshape(signal2[0, max(0,k-round(window_size/2)+1):k],(1,-1)),np.reshape(signal2[:,k],(1,-1)),np.reshape(signal2[0, k+1:min(k+round(window_size/2)-1,signal2.shape[1])],(1,-1)))) #mirroring signal towards the beginning and end
        window=np.hstack((np.reshape(signal2[0, max(0,k-round(window_size/2)+1):k+1],(1,-1)),np.reshape(signal2[:,k],(1,-1)),np.reshape(signal2[0, k+1:min(k+round(window_size/2)-1,signal2.shape[1])],(1,-1)))) #mirroring signal towards the beginning and end
    
        if (k-round(window_size/2))<0:
            window=np.hstack((np.zeros((1,round(window_size/2)-k)),window))
    
    
        if (k+round(window_size/2))>signal2.shape[1]:
            window=np.hstack((window, np.zeros((1,(window_size-window.shape[1])))))
    
        window1=np.diag(window[0,:]);
        tot=np.matmul(window1,P)
        total=np.sum(tot,axis=0);
        total1=np.fliplr(np.reshape(total,(1,-1)))
        total1=total1[0,:]
        I=np.argmax(total1)
        
        beg_last=window[0,0]
        correction=lag[I]
        corrected_curve=np.roll(curve.transpose()[0,:],-correction)
        diff=np.var(corrected_curve-window)
        c_sqi[0,k]=diff;
        
        if c_sqi[0,k]!=0:
            sqi[0,k]=1/c_sqi[0,k]
        else:
            sqi[0,k]=0.014;
    
        if sqi[0,k]<0: #limiting negative SQI values to 0.0001
            sqi[0,k]=0
            
        
    
    


        for k in range(1,signal2.shape[1]):
            window=window[:,1:]
        
        
            if (round(window_size/2)+k-1>signal2.shape[1]):
                window=np.hstack((window, np.zeros((1,(window_size-window.shape[1])))))
            else:
                window=np.hstack((window,np.reshape(signal2[:,round(window_size/2)+k-2],(1,-1))))
            
        
            last_term=window[0,-1];
            new_row=last_term*P[-1,:];
            first_row=beg_last*P[0,:];
            total=total-first_row;
            total=np.roll(total,-1);
            total=total+new_row;
            beg_last=window[0,0];
            total1=np.fliplr(np.reshape(total,(1,-1)));
            total1=total1[0,:]
            I=np.argmax(total1)
            correction=lag[I];
            corrected_curve=np.roll(curve.transpose()[0,:],-correction);
            diff=np.var(corrected_curve-window)
            
            if k>=signal.shape[1]:
                
                c_sqi=np.hstack((c_sqi,np.reshape(diff,(1,-1))))
                if c_sqi[0,k]!=0:
                    sqi=np.hstack((sqi,np.reshape(1/c_sqi[0,k],(1,-1))))
                else:
                    sqi=np.hstack((sqi,np.reshape(0.014,(1,-1))))
            else:
                c_sqi[0,k]=diff
                if c_sqi[0,k]!=0:
                    sqi[0,k]=1/c_sqi[0,k]
                else:
                    sqi[0,k]=0.014;
    
            if sqi[0,k]<0: #limiting negative SQI values to 0.0001
                sqi[0,k]=0
        

        sqi=sqi[:,round((sqi.shape[1]-signal.shape[1])/2)+1:round((sqi.shape[1]+signal.shape[1])/2)+1];
    return sqi

```

```python
def scale_mul(signals,mul,diff):
    
    signal_copy=signals.copy()
    return signal_copy*mul

def shift_add(signals,addi,diff):
    
    signal_copy=signals.copy()
    return signal_copy+addi

class augment_class_simult:
        
    def scale(self):
        if self.both:
            signals_add=self.signals
            signal=np.append(self.signals,signals_add, axis=0) 
            signals_copy=signal.copy()
            
            mul=(1 - -1) * np.random.random_sample(len(signals_copy)) + -1
            aug_signal_1=[scale_mul(signals_copy[i,0:round((signals_copy.shape[1])/2)],mul[i],self.diff) for i in range(len(signals_copy))]
            aug_signal_2=[scale_mul(signals_copy[i,round((signals_copy.shape[1])/2):signals_copy.shape[1]],mul[i],self.diff) for i in range(len(signals_copy))]
            aug_signal=np.append(np.array(aug_signal_1), np.array(aug_signal_2), axis=1)
            
            self.aug=np.append(self.aug,aug_signal, axis=0) 
        else:
            
            if self.diff>self.signals.shape[0]:
                rem=self.signals.shape[0]
                signals=np.append(self.signals,self.signals, axis=0)
                rec=random.sample(range(signals.shape[0]), abs(diff-rem))
                signals_add=signals[rec,:]
            else:
                rec=random.sample(range(self.signals.shape[0]), self.diff)
                signals_add=self.signals[rec,:]
            signal=np.append(self.signals,signals_add, axis=0) 
            signals_copy=signal.copy()
        
            mul=(1 - -1) * np.random.random_sample(len(signals_copy)) + -1
        
            aug_signal=[scale_mul(signals_copy[i,:],mul[i],self.diff) for i in range(len(signals_copy))]
            self.aug=np.append(self.aug,np.array(aug_signal), axis=0) 
    
    def flip(self):
        if self.both:
            signals_add=self.signals
            signal=np.append(self.signals,signals_add, axis=0) 
            signals_copy=signal.copy()
            
            mul=[random.choice([-1,1]) for i in range(len(signals_copy))]
            aug_signal_1=[scale_mul(signals_copy[i,0:round((signals_copy.shape[1])/2)],mul[i],self.diff) for i in range(len(signals_copy))]
            aug_signal_2=[scale_mul(signals_copy[i,round((signals_copy.shape[1])/2):signals_copy.shape[1]],mul[i],self.diff) for i in range(len(signals_copy))]
            
            aug_signal=np.append(np.array(aug_signal_1), np.array(aug_signal_2), axis=1)
            self.aug=np.append(self.aug,aug_signal, axis=0) 
        else:
            
            if self.diff>self.signals.shape[0]:
                rem=self.signals.shape[0]
                signals=np.append(self.signals,self.signals, axis=0)
                rec=random.sample(range(signals.shape[0]), abs(diff-rem))
                signals_add=signals[rec,:]
            else:
                rec=random.sample(range(self.signals.shape[0]), self.diff)
                signals_add=self.signals[rec,:]
            signal=np.append(self.signals,signals_add, axis=0) 
            signals_copy=signal.copy()
        
            mul=[random.choice([-1,1]) for i in range(len(signals_copy))]
            aug_signal=[scale_mul(signals_copy[i,:],mul[i],self.diff) for i in range(len(signals_copy))]
            self.aug=np.append(self.aug,np.array(aug_signal), axis=0) 
        
    
    def cutout(self):
        if self.both:
            signals_add=self.signals
            signal=np.append(self.signals,signals_add, axis=0) 
            signals_copy=signal.copy()
            
            aug_signals=signals_copy.copy()
            indices1=[np.random.randint(10,2200)+np.random.randint(-10,10) for i in range(len(aug_signals))]
            add1=[(150+np.random.randint(-30,30)) for i in range(len(aug_signals))]
            indices2=[]
            for (item1, item2) in zip(indices1, add1):
                indices2.append(item1+item2)
            for i in range(len(aug_signals)):
                aug_signals[i,indices1[i]:indices2[i]]=0 
                
            indices1=[np.random.randint(2520,4800)+np.random.randint(-10,10) for i in range(len(aug_signals))]
            add1=[(150+np.random.randint(-30,30)) for i in range(len(aug_signals))]
            indices2=[]
            for (item1, item2) in zip(indices1, add1):
                indices2.append(item1+item2)
            for i in range(len(aug_signals)):
                aug_signals[i,indices1[i]:indices2[i]]=0 
            self.aug=np.append(self.aug,np.array(aug_signals), axis=0) 
            
            
        else:
            
            if self.diff>self.signals.shape[0]:
                rem=self.signals.shape[0]
                signals=np.append(self.signals,self.signals, axis=0)
                rec=random.sample(range(signals.shape[0]), abs(diff-rem))
                signals_add=signals[rec,:]
            else:
                rec=random.sample(range(self.signals.shape[0]), self.diff)
                signals_add=self.signals[rec,:]
            signal=np.append(self.signals,signals_add, axis=0) 
            signals_copy=signal.copy()
        
            aug_signals=signals_copy.copy()
            indices1=[np.random.randint(10,2200)+np.random.randint(-10,10) for i in range(len(aug_signals))]
            add1=[(150+np.random.randint(-30,30)) for i in range(len(aug_signals))]
            indices2=[]
            for (item1, item2) in zip(indices1, add1):
                indices2.append(item1+item2)
            for i in range(len(aug_signals)):
                aug_signals[i,indices1[i]:indices2[i]]=0 
            self.aug=np.append(self.aug,np.array(aug_signals), axis=0) 
        
        
    
    def shift(self):
        signals_add=self.signals
        signal=np.append(self.signals,signals_add, axis=0) 
        signals_copy=signal.copy()
        
        addi=(3 - -3) * np.random.random_sample(len(signals_copy)) + -3
        aug_signal=[shift_add(signals_copy[i,:],addi[i],self.diff) for i in range(len(signals_copy))]
        self.aug=np.append(self.aug,np.array(aug_signal), axis=0) 
    
    def sine(self):
        SNR_dB=10
        signals_add=self.signals
        signal=np.append(self.signals,signals_add, axis=0) 
        signals_copy=signal.copy()
        
        aug_signal=signals_copy.copy()
        
        if self.both:
            for i in range(len(signals_copy)):
                samples = np.arange(round(signals_copy.shape[1]/2))/125
                f=(1-0) * np.random.random_sample(1)
                nois = np.sin(2 * np.pi * f * samples)
                nois=(nois-np.mean(nois))/np.linalg.norm(nois,2)
                squares=np.linalg.norm(signals_copy[i,0:round(signals_copy.shape[1]/2)],2)**2
                variance=squares/(10**(SNR_dB/10))
                k=(variance - 0) * np.random.random_sample(1) 
                aug_signal[i,0:round(signals_copy.shape[1]/2)]=signals_copy[i,0:round(signals_copy.shape[1]/2)]+np.sqrt(k)*nois
                
                squares=np.linalg.norm(signals_copy[i,round(signals_copy.shape[1]/2):signals_copy.shape[1]],2)**2
                variance=squares/(10**(SNR_dB/10))
                k=(variance - 0) * np.random.random_sample(1) 
                aug_signal[i,round(signals_copy.shape[1]/2):signals_copy.shape[1]]=signals_copy[i,round(signals_copy.shape[1]/2):signals_copy.shape[1]]+np.sqrt(k)*nois
                
            self.aug=np.append(self.aug,np.array(aug_signal), axis=0) 
        else:
            
            for i in range(len(signals_copy)):
                samples = np.arange(signals_copy.shape[1])/125
                f=(1-0) * np.random.random_sample(1)
                nois = np.sin(2 * np.pi * f * samples)
                nois=(nois-np.mean(nois))/np.linalg.norm(nois,2)
                squares=np.linalg.norm(signals_copy[i,:],2)**2
                variance=squares/(10**(SNR_dB/10))
                k=(variance - 0) * np.random.random_sample(1) 
                aug_signal[i,:]=signals_copy[i,:]+np.sqrt(k)*nois
            self.aug=np.append(self.aug,np.array(aug_signal), axis=0) 
    
    def square(self):
        SNR_dB=13
        signals_add=self.signals
        signals=np.append(self.signals,signals_add, axis=0) 
        signals_copy=signals.copy()
        
        aug_signal=signals_copy.copy()
        if self.both:
            for i in range(len(signals_copy)):
                samples = np.arange(round(signals_copy.shape[1]/2))/125
                f=(1-0) * np.random.random_sample(1)
                nois = signal.square(2 * np.pi * f * samples)
                nois=(nois-np.mean(nois))/np.linalg.norm(nois,2)
                squares=np.linalg.norm(signals_copy[i,0:round(signals_copy.shape[1]/2)],2)**2
                variance=squares/(10**(SNR_dB/10))
                k=(variance - 0) * np.random.random_sample(1) 
                aug_signal[i,0:round(signals_copy.shape[1]/2)]=signals_copy[i,0:round(signals_copy.shape[1]/2)]+np.sqrt(k)*nois
                
                squares=np.linalg.norm(signals_copy[i,round(signals_copy.shape[1]/2):signals_copy.shape[1]],2)**2
                variance=squares/(10**(SNR_dB/10))
                k=(variance - 0) * np.random.random_sample(1)
                aug_signal[i,round(signals_copy.shape[1]/2):signals_copy.shape[1]]=signals_copy[i,round(signals_copy.shape[1]/2):signals_copy.shape[1]]+np.sqrt(k)*nois
                
            self.aug=np.append(self.aug,np.array(aug_signal), axis=0)
            
        else:
            for i in range(len(signals_copy)):
                samples = np.arange(signals_copy.shape[1])/125
                f=(1-0) * np.random.random_sample(1)
                nois = signal.square(2 * np.pi * f * samples)
                nois=(nois-np.mean(nois))/np.linalg.norm(nois,2)
                squares=np.linalg.norm(signals_copy[i,:],2)**2
                variance=squares/(10**(SNR_dB/10))
                k=(variance - 0) * np.random.random_sample(1) 
                aug_signal[i,:]=signals_copy[i,:]+np.sqrt(k)*nois
            self.aug=np.append(self.aug,np.array(aug_signal), axis=0) 

    def white(self):
        signals_add=self.signals
        signal=np.append(self.signals,signals_add, axis=0) 
        signals_copy=signal.copy()
        
        aug_signal=signals_copy.copy()
        SNR_dB=20
        
        if self.both:
            for i in range(len(signals_copy)):
                
                squares=np.linalg.norm(signals_copy[i,0:round(signals_copy.shape[1]/2)],2)**2
                variance=squares/(10**(SNR_dB/10))
                k=(variance - 0) * np.random.random_sample(1) 
                nois=np.random.normal(size=round(signals_copy.shape[1]/2))
                nois=(nois-np.mean(nois))/np.linalg.norm(nois,2)
                aug_signal[i,0:round(signals_copy.shape[1]/2)]=signals_copy[i,0:round(signals_copy.shape[1]/2)]+np.sqrt(k)*nois
                
                squares=np.linalg.norm(signals_copy[i,round(signals_copy.shape[1]/2):signals_copy.shape[1]],2)**2
                variance=squares/(10**(SNR_dB/10))
                k=(variance - 0) * np.random.random_sample(1) 
                nois=np.random.normal(size=round(signals_copy.shape[1]/2))
                nois=(nois-np.mean(nois))/np.linalg.norm(nois,2)
                aug_signal[i,round(signals_copy.shape[1]/2):signals_copy.shape[1]]=signals_copy[i,round(signals_copy.shape[1]/2):signals_copy.shape[1]]+np.sqrt(k)*nois
                
            self.aug=np.append(self.aug,np.array(aug_signal), axis=0)
            
        else:
            for i in range(len(signals_copy)):
                squares=np.linalg.norm(signals_copy[i,:],2)**2
                variance=squares/(10**(SNR_dB/10))
                k=(variance - 0) * np.random.random_sample(1) 
                nois=np.random.normal(size=len(signals_copy[i,:]))
                nois=(nois-np.mean(nois))/np.linalg.norm(nois,2)
                aug_signal[i,:]=signals_copy[i,:]+np.sqrt(k)*nois

            self.aug=np.append(self.aug,np.array(aug_signal), axis=0) 

    def partial_sine(self):
        SNR_dB=5
        signals_add=self.signals
        signal=np.append(self.signals,signals_add, axis=0) 
        signals_copy=signal.copy()
        
        aug_signal=signals_copy.copy()
        
        if self.both:
            indices1=[np.random.randint(10,2200)+np.random.randint(-10,10) for i in range(len(aug_signal))]
            add1=[(100+np.random.randint(-30,30)) for i in range(len(aug_signal))]
            indices2=[]
            for (item1, item2) in zip(indices1, add1):
                indices2.append(item1+item2)
            for i in range(len(signals_copy)):
                samples = np.linspace(indices1[i],indices2[i],(indices2[i]-indices1[i]+1))/125
                f=(1-0) * np.random.random_sample(1)
                nois = np.sin(2 * np.pi * f * samples)
                nois=(nois-np.mean(nois))/np.linalg.norm(nois,2)
                squares=np.linalg.norm(signals_copy[i,indices1[i]:indices2[i]+1],2)**2
                variance=squares/(10**(SNR_dB/10))
                k=(variance - 0) * np.random.random_sample(1) 
                
                aug_signal[i,indices1[i]:indices2[i]+1]=signals_copy[i,indices1[i]:indices2[i]+1]+np.sqrt(k)*nois
                
            indices1=[np.random.randint(2520,4800)+np.random.randint(-10,10) for i in range(len(aug_signal))]
            add1=[(100+np.random.randint(-30,30)) for i in range(len(aug_signal))]
            indices2=[]
            for (item1, item2) in zip(indices1, add1):
                indices2.append(item1+item2)
            for i in range(len(signals_copy)):
                samples = np.linspace(indices1[i],indices2[i],(indices2[i]-indices1[i]+1))/125
                f=(1-0) * np.random.random_sample(1)
                nois = np.sin(2 * np.pi * f * samples)
                nois=(nois-np.mean(nois))/np.linalg.norm(nois,2)
                squares=np.linalg.norm(signals_copy[i,indices1[i]:indices2[i]+1],2)**2
                variance=squares/(10**(SNR_dB/10))
                k=(variance - 0) * np.random.random_sample(1) 
                
                
                aug_signal[i,indices1[i]:indices2[i]+1]=signals_copy[i,indices1[i]:indices2[i]+1]+np.sqrt(k)*nois
            self.aug=np.append(self.aug,np.array(aug_signal), axis=0)
            
            
        else:
            indices1=[np.random.randint(10,2200)+np.random.randint(-10,10) for i in range(len(aug_signal))]
            add1=[(100+np.random.randint(-30,30)) for i in range(len(aug_signal))]
            indices2=[]
            for (item1, item2) in zip(indices1, add1):
                indices2.append(item1+item2)
            for i in range(len(signals_copy)):
                samples = np.linspace(indices1[i],indices2[i],(indices2[i]-indices1[i]+1))/125
                f=(1-0) * np.random.random_sample(1)
                nois = np.sin(2 * np.pi * f * samples)
                nois=(nois-np.mean(nois))/np.linalg.norm(nois,2)
                squares=np.linalg.norm(signals_copy[i,indices1[i]:indices2[i]+1],2)**2
                variance=squares/(10**(SNR_dB/10))
                k=(variance - 0) * np.random.random_sample(1) 
                aug_signal[i,indices1[i]:indices2[i]+1]=signals_copy[i,indices1[i]:indices2[i]+1]+np.sqrt(k)*nois
            self.aug=np.append(self.aug,np.array(aug_signal), axis=0) 


    def partial_square(self):
        SNR_dB=5
        signals_add=self.signals
        signals=np.append(self.signals,signals_add, axis=0) 
        signals_copy=signals.copy()
        
        aug_signal=signals_copy.copy()
        
        if self.both:
            indices1=[np.random.randint(10,2200)+np.random.randint(-10,10) for i in range(len(aug_signal))]
            add1=[(100+np.random.randint(-30,30)) for i in range(len(aug_signal))]
            indices2=[]
            for (item1, item2) in zip(indices1, add1):
                indices2.append(item1+item2)
            for i in range(len(signals_copy)):
                samples = np.linspace(indices1[i],indices2[i],(indices2[i]-indices1[i]+1))/125
                f=(1-0) * np.random.random_sample(1)
                nois = signal.square(2 * np.pi * f * samples)
                nois=(nois-np.mean(nois))/np.linalg.norm(nois,2)
                squares=np.linalg.norm(signals_copy[i,indices1[i]:indices2[i]+1],2)**2
                variance=squares/(10**(SNR_dB/10))
                k=(variance - 0) * np.random.random_sample(1) 
                aug_signal[i,indices1[i]:indices2[i]+1]=signals_copy[i,indices1[i]:indices2[i]+1]+np.sqrt(k)*nois
                
            indices1=[np.random.randint(2520,4800)+np.random.randint(-10,10) for i in range(len(aug_signal))]
            add1=[(100+np.random.randint(-30,30)) for i in range(len(aug_signal))]
            indices2=[]
            for (item1, item2) in zip(indices1, add1):
                indices2.append(item1+item2)
            for i in range(len(signals_copy)):
                samples = np.linspace(indices1[i],indices2[i],(indices2[i]-indices1[i]+1))/125
                f=(1-0) * np.random.random_sample(1)
                nois = signal.square(2 * np.pi * f * samples)
                nois=(nois-np.mean(nois))/np.linalg.norm(nois,2)
                squares=np.linalg.norm(signals_copy[i,indices1[i]:indices2[i]+1],2)**2
                variance=squares/(10**(SNR_dB/10))
                k=(variance - 0) * np.random.random_sample(1) 
                aug_signal[i,indices1[i]:indices2[i]+1]=signals_copy[i,indices1[i]:indices2[i]+1]+np.sqrt(k)*nois
            self.aug=np.append(self.aug,np.array(aug_signal), axis=0)
            
        else:
        
            indices1=[np.random.randint(10,2200)+np.random.randint(-10,10) for i in range(len(aug_signal))]
            add1=[(100+np.random.randint(-30,30)) for i in range(len(aug_signal))]
            indices2=[]
            for (item1, item2) in zip(indices1, add1):
                indices2.append(item1+item2)
            for i in range(len(signals_copy)):
                samples = np.linspace(indices1[i],indices2[i],(indices2[i]-indices1[i]+1))/125
                f=(1-0) * np.random.random_sample(1)
                nois = signal.square(2 * np.pi * f * samples)
                nois=(nois-np.mean(nois))/np.linalg.norm(nois,2)
                squares=np.linalg.norm(signals_copy[i,indices1[i]:indices2[i]+1],2)**2
                variance=squares/(10**(SNR_dB/10))
                k=(variance - 0) * np.random.random_sample(1) 
                aug_signal[i,indices1[i]:indices2[i]+1]=signals_copy[i,indices1[i]:indices2[i]+1]+np.sqrt(k)*nois
            self.aug=np.append(self.aug,np.array(aug_signal), axis=0) 


    def partial_white(self):
        SNR_dB=7
        signals_add=self.signals
        signal=np.append(self.signals,signals_add, axis=0) 
        signals_copy=signal.copy()
        
        aug_signal=signals_copy.copy()
        
        if self.both:
            indices1=[np.random.randint(10,2200)+np.random.randint(-10,10) for i in range(len(aug_signal))]
            add1=[(100+np.random.randint(-30,30)) for i in range(len(aug_signal))]
            indices2=[]
            for (item1, item2) in zip(indices1, add1):
                indices2.append(item1+item2)
            for i in range(len(signals_copy)):
                samples = np.linspace(indices1[i],indices2[i],(indices2[i]-indices1[i]+1))/125
                f=(1-0) * np.random.random_sample(1)
                nois = np.random.normal(size=(indices2[i]-indices1[i]+1))
                nois=(nois-np.mean(nois))/np.linalg.norm(nois,2)
                squares=np.linalg.norm(signals_copy[i,indices1[i]:indices2[i]+1],2)**2
                variance=squares/(10**(SNR_dB/10))
                k=(variance - 0) * np.random.random_sample(1) 
                aug_signal[i,indices1[i]:indices2[i]+1]=signals_copy[i,indices1[i]:indices2[i]+1]+np.sqrt(k)*nois
                
            indices1=[np.random.randint(2520,4800)+np.random.randint(-10,10) for i in range(len(aug_signal))]
            add1=[(100+np.random.randint(-30,30)) for i in range(len(aug_signal))]
            indices2=[]
            for (item1, item2) in zip(indices1, add1):
                indices2.append(item1+item2)
            for i in range(len(signals_copy)):
                samples = np.linspace(indices1[i],indices2[i],(indices2[i]-indices1[i]+1))/125
                f=(1-0) * np.random.random_sample(1)
                nois = np.random.normal(size=(indices2[i]-indices1[i]+1))
                nois=(nois-np.mean(nois))/np.linalg.norm(nois,2)
                squares=np.linalg.norm(signals_copy[i,indices1[i]:indices2[i]+1],2)**2
                variance=squares/(10**(SNR_dB/10))
                k=(variance - 0) * np.random.random_sample(1) 
                aug_signal[i,indices1[i]:indices2[i]+1]=signals_copy[i,indices1[i]:indices2[i]+1]+np.sqrt(k)*nois
            self.aug=np.append(self.aug,np.array(aug_signal), axis=0)
            
        else:
        
            indices1=[np.random.randint(10,2200)+np.random.randint(-10,10) for i in range(len(aug_signal))]
            add1=[(100+np.random.randint(-30,30)) for i in range(len(aug_signal))]
            indices2=[]
            for (item1, item2) in zip(indices1, add1):
                indices2.append(item1+item2)
            for i in range(len(signals_copy)):
                nois=np.random.normal(size=(indices2[i]-indices1[i]+1))
                nois=(nois-np.mean(nois))/np.linalg.norm(nois,2)
                squares=np.linalg.norm(signals_copy[i,indices1[i]:indices2[i]+1],2)**2
                variance=squares/(10**(SNR_dB/10))
                k=(variance - 0) * np.random.random_sample(1) 
                aug_signal[i,indices1[i]:indices2[i]+1]=signals_copy[i,indices1[i]:indices2[i]+1]+np.sqrt(k)*nois
            self.aug=np.append(self.aug,np.array(aug_signal), axis=0) 


    def fir_low(self):
        signals_add=self.signals
        signals=np.append(self.signals,signals_add, axis=0) 
        signals_copy=signals.copy()
        
        aug_signal=signals_copy.copy()
        nyq_rate = 125 / 2
        # Width of the roll-off region.
        width = 5 / nyq_rate
        # Attenuation in the stop band.
        ripple_db = 12.0
        num_of_taps, beta = signal.kaiserord(ripple_db, width)
        if num_of_taps % 2 == 0:
            num_of_taps = num_of_taps + 1
        # Cut-off frequency.
        cutoff_hz = 55
        # Estimate the filter coefficients.
        taps = signal.firwin(num_of_taps, cutoff_hz/nyq_rate, window=('kaiser', beta), pass_zero=True)
        
        if self.both:
            for i in range(len(signals_copy)):
                aug_signal[i,0:round(signals_copy.shape[1]/2)]=signal.convolve(signals_copy[i,0:round(signals_copy.shape[1]/2)], taps, mode='same', method='auto')
                aug_signal[i,round(signals_copy.shape[1]/2):signals_copy.shape[1]]=signal.convolve(signals_copy[i,round(signals_copy.shape[1]/2):signals_copy.shape[1]], taps, mode='same', method='auto')
            self.aug=np.append(self.aug,np.array(aug_signal), axis=0)
        else:
            for i in range(len(signals_copy)):
                aug_signal[i,:]=signal.convolve(signals_copy[i,:], taps, mode='same', method='auto')
            self.aug=np.append(self.aug,np.array(aug_signal), axis=0)


    def fir_high(self):
        signals_add=self.signals
        signals=np.append(self.signals,signals_add, axis=0) 
        signals_copy=signals.copy()
        
        aug_signal=signals_copy.copy()
        
        nyq_rate = 125 / 2
        # Width of the roll-off region.
        width = 0.5 / nyq_rate
        # Attenuation in the stop band.
        ripple_db = 12.0
        num_of_taps, beta = signal.kaiserord(ripple_db, width)
        if num_of_taps % 2 == 0:
            num_of_taps = num_of_taps + 1
        # Cut-off frequency.
        cutoff_hz = 1
        # Estimate the filter coefficients.
        taps = signal.firwin(num_of_taps, cutoff_hz/nyq_rate, window=('kaiser', beta), pass_zero=False)
        if self.both:
            for i in range(len(signals_copy)):
                aug_signal[i,0:round(signals_copy.shape[1]/2)]=signal.convolve(signals_copy[i,0:round(signals_copy.shape[1]/2)], taps, mode='same', method='auto')
                aug_signal[i,round(signals_copy.shape[1]/2):signals_copy.shape[1]]=signal.convolve(signals_copy[i,round(signals_copy.shape[1]/2):signals_copy.shape[1]], taps, mode='same', method='auto')
            self.aug=np.append(self.aug,np.array(aug_signal), axis=0)
        else:
            for i in range(len(signals_copy)):
                aug_signal[i,:]=signal.convolve(signals_copy[i,:], taps, mode='same', method='auto')
            self.aug=np.append(self.aug,np.array(aug_signal), axis=0) 
    
    def augment(self,signals, label, diff, both=False,add_noise=True, rep=0):
        if add_noise:
            if diff>signals.shape[0]:
                rem=signals.shape[0]
                signals=np.append(signals,signals, axis=0)
                rec=random.sample(range(signals.shape[0]), abs(diff-rem))
                signals_add=signals[rec,:]
            else:
                rec=random.sample(range(signals.shape[0]), diff)
                signals_add=signals[rec,:]
            new_signals=np.append(signals,signals_add, axis=0) 
            self.orig_signals=new_signals
            self.label=label
            #self.aug=random.samplenew_signals.copy()
            
            rec=random.sample(range(new_signals.shape[0]),round(new_signals.shape[0]/50))
            self.signals=new_signals[rec,:]
            self.aug=self.signals
            self.diff=diff
            self.both=both
        
            print('scale')
            self.scale()
            print('flip')
            self.flip()
            print('cutout')
            self.cutout()
            print('shift')
            self.shift()
            print('sine')
            self.sine()
            print('square')
            self.square()
            print('white')
            self.white()
            print('partial_sine')
            self.partial_sine()
            print('partial_square')
            self.partial_square()
            print('partial_white')
            self.partial_white()
            print('fir_low')
            self.fir_low()
            print('fir_high')
            self.fir_high()
            
            self.aug=np.append(self.orig_signals,self.aug, axis=0)

            if self.label==0:
                labels=np.zeros((self.aug.shape[0]))
            if self.label==1:
                labels=np.ones((self.aug.shape[0]))
            
        else:
            self.label=label
            if diff>signals.shape[0]:
                rem=signals.shape[0]
                signals=np.append(signals,signals, axis=0)
                rec=random.sample(range(signals.shape[0]), abs(diff-rem))
                signals_add=signals[rec,:]
                
            else:
                rec=random.sample(range(signals.shape[0]), diff)
                
                signals_add=signals[rec,:]
                
            
            signals=np.append(signals,signals_add, axis=0) 
            
            self.signals=signals
            self.label=label
            self.aug=self.signals.copy()
            
            for i in range(rep):
                self.aug=np.append(self.aug,signals, axis=0)
            if self.label==0:
                labels=np.zeros((self.aug.shape[0]))
            if self.label==1:
                labels=np.ones((self.aug.shape[0]))
        return self.aug, labels
```
